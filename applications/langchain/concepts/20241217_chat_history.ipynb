## Chat History Management

This document outlines how chat history is managed within chat models, focusing on its structure and importance for maintaining context.

### Core Concepts

*   **Chat history** is a record of the conversation between the user and the chat model, essential for maintaining context and state.
*   It is a sequence of **messages**, each associated with a specific **role**.

### Conversation Patterns

*   Conversations typically start with a **system message** to set the context.
*   This is followed by a **user message** containing user input, and then an **assistant message** with the model's response.
*   Two main patterns of alternating messages are:
    *   **User** and **assistant** for back-and-forth conversations.
    *  **Assistant** and **tool messages** when the assistant invokes tools.

### Managing Chat History

*   Chat models have a **maximum input size**, requiring careful management of chat history to avoid exceeding the context window.
*   Maintaining a **correct conversation structure** is essential when processing chat history.
*   Key guidelines for managing chat history:
    *   The first message should be either a **user** or **system** message, followed by a **user** and then an **assistant** message.
    *   The last message should be either a **user** or a **tool** message containing the result of a tool call.
    *   A **tool** message must follow an **assistant** message that requested the tool.

### Importance of Correct Conversation Structure

*   Understanding correct conversation structure is essential for implementing **memory** in chat models.
