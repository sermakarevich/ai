# Multi-agent Systems in LangGraph

The article discusses various architectures and communication methods for building multi-agent systems in LangGraph, summarized below:

## Multi-agent Architectures

- **Network**: Each agent can communicate with every other agent, allowing any agent to decide which one to call next. 

- **Supervisor**: A single supervisor agent communicates with all other agents, making decisions on which agent should be called next. 

- **Supervisor (Tool-Calling)**: A variation where individual agents are represented as tools, and the supervisor uses a tool-calling LLM to decide which agent tool to call and what arguments to pass. 

- **Hierarchical**: A multi-agent system with multiple layers of supervisors, allowing for more complex control flows.

- **Custom Multi-agent Workflow**: Agents communicate with only a subset of other agents, with parts of the flow being deterministic and some agents deciding which others to call next. 

### Handoffs

- **Handoffs**: A pattern where one agent hands off control to another, specifying the destination agent and the information to pass along.

- **Handoffs as Tools**: In tool-calling agents, handoffs can be wrapped in tool calls, allowing the supervisor to manage control flow and state updates effectively. 

## Communication Between Agents

- **Graph State vs. Tool Calls**: Agents can communicate via shared graph state or through tool calls, depending on the architecture and requirements. 

- **Different State Schemas**: When agents have different state schemas, it's essential to manage state transformations to ensure seamless communication. 

- **Shared Message List**: Agents can share a message list to maintain context, either by sharing the full history or just the final result, facilitating effective collaboration.

These concepts provide a framework for designing and implementing complex multi-agent systems in LangGraph, enabling modularity, specialization, and controlled communication.
